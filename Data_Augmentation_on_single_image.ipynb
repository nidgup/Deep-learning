```python
from tensorflow.keras.preprocessing import image
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
img = image.load_img('/content/sample_data/th.jpg', target_size = (200,200))
img
```

```python
type(img)  #type pil image not numpy
```

```python
img = image.img_to_array(img)
type(img)  #converted into numpy array - so we can send it as a batch
```

```python
plt.imshow(img) #pixel values
```

```python
img
```

```python
datagen = ImageDataGenerator(
        rotation_range=40,
        width_shift_range=0.2,
        height_shift_range=0.2,
        rescale=1./255,
        shear_range=0.2,
        zoom_range=0.2,
        horizontal_flip=True,
        fill_mode='reflect')


#while shifting fill pixels with nearest pixels or reflect or constant which is black
```

```python
input_batch = img.reshape(1,200,200,3)
input_batch.shape
```

```python
i = 0

for output in datagen.flow(input_batch, batch_size =1, save_to_dir = '/content/sample_data/train/aug'):
  i = i+1
  if i ==10:
    break
```
