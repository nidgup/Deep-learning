```python
from keras.preprocessing import image
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image_path = '/content/animal-dog-pet-248307.jpg'
img =image.load_img(image_path)
```

```python
type(img)  #this is a pil iamge not an array
```

```python
img_arr = image.img_to_array(img)
type(img_arr)
```

```python
img_arr
```

```python
img_arr.shape
```

```python
img_arr_resize = cv2.resize(img_arr,(200,200))
img_arr_resize.shape
```

```python
grayscale = cv2.cvtColor(img_arr_resize, cv2.COLOR_RGB2GRAY)
grayscale.shape
```

```python
plt.imshow(grayscale)
```

```python
#convert in batch

grayscale_batch = np.reshape(grayscale, (1,200,200,1))
grayscale_batch.shape
```

```python
#saving image

cv2.imwrite('grayscale.jpg',grayscale)
```
